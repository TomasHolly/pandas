Joining 

```python
df1 = pd.DataFrame(
    data = {
        "first":[1,1,1,1],
        "second":[1,1,1,1],
        "third":[1,2,1,2],
               
    }, index= [1,2,3,4])

df2 = pd.DataFrame(
    data = {
        "first":[2,2,2,2],
        "second":[2,2,2,2],
        "fourth":[1,3,1,3],
        
    },index = [3,4,5,6])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2], join="outer") # Pandas zobral tieti dva dataframy a spojil ich jeden na druhého.
```

```python
pd.concat([df1,df2], join="outer", axis = 1)
```

```python
pd.concat([df1,df2], join="inner", axis = 1) # vrátilo nám to iba riadky 3 a 4, lebo boli špecifikované v obidvoch dataframov
```

```python
pd.concat([df1,df2], join="inner", axis = 0)
```

V pandase môžeme joinovať priamo na dataframe:

```python
df1.join(df2, rsuffix="second")
```

```python
df1.join(df2, lsuffix="second")
```

```python
df1.join(df2, lsuffix="second", how = "right")
```

```python
pd.concat([df1,df2], axis = 1)
```
